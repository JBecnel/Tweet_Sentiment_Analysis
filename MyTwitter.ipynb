```python
import os
import json
def appendTweets(filename, tweets_to_write):
    if os.path.exists(filename):
        with open(filename, 'r+') as file:
            # First we load existing data into a dict.
            file_data = json.load(file)
            # Join new_dat3a with file_data
            #for t in tweets_to_write:
              ##  file_data += t
            file_data += tweets_to_write
            # Sets file's current position at offset.
            file.seek(0)
            # convert back to json.
            json.dump(file_data, file, indent = 1)
    else:
        with open(filename, 'w+') as file:
            # convert back to json.
            json.dump(tweets_to_write, file, indent = 1)
```

```python
filename = 'my_precious_tweets_on_%s.json'%'test'
appendTweets(filename,json_tweets)
```
