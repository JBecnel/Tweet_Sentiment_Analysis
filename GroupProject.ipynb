## Create json file for tweets
Below we create a json file for the tweets we collected. This code is modfied from the one given in the workshop.

```python
import os
import json
def appendTweets(filename, tweets_to_write):
    if os.path.exists(filename):
        with open(filename, 'r+') as file:
            # First we load existing data into a dict.
            file_data = json.load(file)
            # Join new_dat3a with file_data
            #for t in tweets_to_write:
              ##  file_data += t
            file_data += tweets_to_write
            # Sets file's current position at offset.
            file.seek(0)
            # convert back to json.
            json.dump(file_data, file, indent = 1)
    else:
        with open(filename, 'w+') as file:
            # convert back to json.
            json.dump(tweets_to_write, file, indent = 1)
```

```python
filename = 'my_precious_tweets_on_%s.json'%'BF42'
appendTweets(filename,json_tweets)
```

## Adding a numerical score
Below we are adding a numerical score for the sentiment. Many tweets have one of three classifications. We are only interested in positive (or not positive).
Thus, we give a score of 1 for positive and 0 for not positive.

```python
def convert_to_num(sentiment):
    if sentiment.lower() == "positive":
        return 1
    elif sentiment.lower() == 'negative':
        return 0  # or -1 if you want 3 classes
    else:
        return 0
```

```python
airline_df['score'] = airline_df['airline_sentiment'].map(convert_to_num)
airline_df.head()
```

```python
### WARNING: EMOTICONS IS A LIST ABOVE...click the .... ###
import re
import emoji
# this will strip URLs from text
url_pattern = re.compile(r'https?://\S+|www\.\S+')
def remove_urls(text):        
    return url_pattern.sub(r'', text)

# this will convert emoticons to text
def convert_emoticons(text):    
    for emot in EMOTICONS:
        text = re.sub(u'('+emot+')', "_".join(EMOTICONS[emot].replace(",","").split()), text)
    return text

def convert_emojis(text):    
    new_text = emoji.demojize(text, delimiters=("", ""))  # 'game is on fire fire
    return new_text

def remove_punctuatoin(text):
    return re.sub(r'[^\w\s]','', text)

def clean_text(text):
    text = str(text)
    text = remove_urls(text)              # remove urls
    text = convert_emoticons(text)        # convert emoticones to text
    text = convert_emojis(text)           # convert emojis to text
    text = remove_puncutatoin(text)
    text = text.lower()
    text = text.replace('-', ' ')
    text = text.replace('_', ' ')
    return text
```

# Training a Model
Here we will train a model on our data set.

```python
import sys
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn import metrics

# split the dataset in training and test set:
x_train, x_test, y_train, y_test = train_test_split(
    combined_df['clean'], combined_df['score'], test_size=0.25, random_state=None)
```

```python
# TASK: Build a vectorizer / classifier pipeline that filters out tokens
# that are too rare or too frequent
pipeline = Pipeline([
    ('vect', TfidfVectorizer(min_df=3, max_df=0.95, stop_words='english')),
    ('clf', SGDClassifier(loss='hinge', penalty='l2',
                           random_state=42,
                           max_iter=5, tol=None))
    #('clf', RandomForestClassifier(random_state=42))
    #('clf', LinearSVC(C=1000)),
])

# TASK: Build a grid search to find out whether unigrams or bigrams are
# more useful.
# Fit the pipeline on the training set using grid search for the parameters
parameters = {
    'vect__ngram_range': [(1, 1), (1, 2), (1,3)],
    'vect__use_idf': (True, False),
    'clf__alpha': (1e-2, 1e-3),
    #'clf__n_estimators': [200, 500],
    #'clf__max_features': ['auto', 'sqrt', 'log2'],
    #'clf__max_depth' : [4,5,6,7,8],
    #'clf__criterion' :['gini', 'entropy']
}
grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1)
grid_search.fit(x_train, y_train)

# TASK: print the cross-validated scores for the each parameters set
# explored by the grid search
print("Grid Search Results: ")
print(grid_search.cv_results_)
print("----------------------------------")

# TASK: Predict the outcome on the testing set and store it in a variable
# named y_predicted
y_predicted = grid_search.predict(x_test)

# Print the classification report
print("Classification Report")
print(metrics.classification_report(y_test, y_predicted,
                                    target_names=['not-positive',  'positive']))
print("------------------------------------")
# Print and plot the confusion matrix
cm = metrics.confusion_matrix(y_test, y_predicted)
print(cm)

import matplotlib.pyplot as plt
plt.matshow(cm)
plt.show()
```
